```python
#Looking at the data it is clear that there are no categorical values, but crosschecking to see if there are more
#than 5 unique value in each column.
def num_cat(df):
    num=[]
    cat=[]
    count=df.nunique()
    for i in df.columns:
        if count[i]>5:
            num.append(i)
        else:
            cat.append(i)
    return(num,cat)
```

```python
#The above plot shows outliers, lets be sure to find the same manually
#finding Outliers using z-score technique. if z>3 it is supposed to be an outlier
def z_outlier(df):
    df_z=df
    z=np.abs(stats.zscore(df_z))
    a=int(df_z.size)
    df_z = df_z[(z>3).all(axis=1)]
    print('Percent of data retained = '+ str(int(df_z.size)/a))
    return(df_z)
```

```python
z_outlier(x)
```

```python
def std_scalar(df):
    ss=StandardScaler()
    x=df.to_numpy()
    x1=pd.DataFrame(ss.fit_transform(x),columns=df.columns)
    return(x1)
```

<h2><font color='red'> Using Lasso-L1 Reg technique</font></h2>


```python
Lasscv=LassoCV(alphas=None,max_iter=100,normalize=True)
Lasscv.fit(x_train,y_train)
```

<h2><font color='red'> Using RidgeCV Reg</font></h2>

```python
ridge_model=Ridge(alpha=ridgecv.alpha_)
ridge_model.fit(x_train, y_train)
```

<h2><font color='red'>Finding the Root Mean Square Values</font></h2>

```python
y_pred_by_lasso =  lasso_reg.predict(x_test)
y_pred_by_ridge = ridge_model.predict(x_test)
y_pred_by_LR=lr.predict(x_test)

mse_lasso=mean_squared_error(y_test,y_pred_by_lasso)
mse_ridge=mean_squared_error(y_test,y_pred_by_ridge)
mse_LR=mean_squared_error(y_test,y_pred_by_LR)

print(f'Logistic Reg: {np.sqrt(mse_LR)}')
print(f'Lasso RMSE: {np.sqrt(mse_lasso)}')
print(f'Ridge RMSE: {np.sqrt(mse_ridge)}')
```
